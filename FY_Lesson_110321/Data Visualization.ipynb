# Data Visualization
An important part of data science is data visualization you use it both when starting the coding process as well as when you are writing about your results. This lesson will walk through different types of graphs, how to make them, and when they are appropriate to use. The general categories of data visualization to be covered are: distribution, correlation, and evolution. The main libraries that are going to be used are matplotlib and seaborn.

## Distribution Graphs
The first two graphs are a density plot and a histogram. These graphs can be used in all the same scenarios and as used to show the distribution of a numerical variable. Histograms are created by putting all the data into distinct bins with the height of the bars being the number of observations in each bin. Density plots are continous plots of a numerical variable (absent of bins). Lets start with a basic histogram of sepal length from the iris dataset. 

```python
### NEW CODE ###
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = sns.load_dataset('iris')
plt.hist(data['sepal_length'])
```

This is a little basic and the viewer is left in the dark about what this data represents. To solve this we should always add axis labels and a title to our graphs. We are also going to bold the title using the "fontweight" parameter, but this can be done for axis labels as well. 

```python
plt.hist(data['sepal_length'])
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Iris Sepal Length', fontweight = 'bold')
plt.show()
```

Blue is a little generic so let use UT burnt orange instead. Matplotlib allows us to control the color of the plot using a existing colors in matplotlib as well as hex codes. The UT hex code is #bf5700 so we will use that for our future graphs. Lets also add some gridlines so that numerical values can be more clearly associated with their frequency bars.

This is pretty basic so lets add some labels. Also we are adding the "showmeans" parameter which places the mean on the boxplot. This can be very useful for visualization because it allows the plot to better demonstrate the skew and kurtosis of the data.

```python
plt.hist(data['sepal_length'], color = '#bf5700')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Iris Sepal Length', fontweight = 'bold')
plt.grid(True)
plt.show()
```

Boxplots are useful for comparing the distribution and range of variables as well. We can segment these boxplots to split along a categorical variable too. 

```python
plt.boxplot(data['sepal_width'])
plt.show()
```

```python
plt.boxplot(data['sepal_width'], showmeans = True)
plt.ylabel('Sepal Width')
plt.title('Distribution of Iris Sepal Width', fontweight = 'bold')
plt.show()
```

Scatterplots allow the user to look at the correlation between two (or three) variables. We can use scatterplots to look at the general linearity of a dataset. 

```python
plt.scatter(data['sepal_length'], data['petal_length'])
```

This graph communications the correlation, but we would not want to put it into a report. Lets add gridlines, titles, and change the color of the graph to burnt orange.

```python
plt.scatter(data['sepal_length'], data['petal_length'], color = '#bf5700')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Correlation Between Sepal & Petal Length')
plt.grid(True)
plt.show()
```

Now we see a good display of the data we can also alter some other aspects of the display. I prefer to have smaller points on the graph for data to make it cleaner and less clunky.

```python
plt.scatter(data['sepal_length'], data['petal_length'], color = '#bf5700', s = 20)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Correlation Between Sepal & Petal Length')
plt.grid(True)
plt.show()
```

Often times when you have a scatterplot you will also want to highlight outliers or insert a trendline to visually demonstrate the analysis that you have done. Lets start by highlighting the grouping of data in the bottom lefthand corner that does not follow the general trend of the rest of the data from the iris dataset.

When highlighting a graph you use the data's coordinates to position the highlighting. Here we are highlighing from 4.0 to 6.0 using yellow with no outline (linewidth = 0) as well as a lower alpha so the data still is visible. 

```python
plt.scatter(data['sepal_length'], data['petal_length'], color = '#bf5700', s = 20)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Correlation Between Sepal & Petal Length')
plt.grid(True)
plt.axvspan(4.0, 6.0, color = 'yellow', alpha = 0.2, linewidth = 0)
plt.show()
```

We can also see a trend in the data as well so lets add a trendline to the data so that our "analysis" can be better visualized. Right now the trendline will be drawn arbitrarly, but we will learn how to find a line of best fit using OSL and a variety of other methods in the future. The trendline we are adding will go from (4.0, 2.0) to (8.0, 7.0). In the ploting function we also specific the line type will be a dashed with a width of 1 in blue. 

```python
plt.scatter(data['sepal_length'], data['petal_length'], color = '#bf5700', s = 20)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Correlation Between Sepal & Petal Length')
plt.grid(True)
plt.axvspan(4.0, 6.0, color = 'yellow', alpha = 0.2, linewidth = 0)
plt.plot([4.0, 8.0], [2.0, 7.0], 'k--', linewidth = 1, color = 'blue')
plt.show()
```

Above we gave matplotlib the points in which to graph our line. We can also determine this use numpy. 

```python
plt.scatter(data['sepal_length'], data['petal_length'], color = '#bf5700', s = 20)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Correlation Between Sepal & Petal Length')
plt.grid(True)


slope, intercept = np.polyfit(data['sepal_length'], data['petal_length'], 1)
abline_values = [slope * i + intercept for i in data['sepal_length']]
plt.plot(data['sepal_length'], abline_values, 
         linestyle = '-', 
         color = 'red')
plt.show()
```

```python
# we just noticed that there is a large amount of error near the central subset of the data; lets adjust
plt.scatter(data['sepal_length'], data['petal_length'], color = '#bf5700', s = 20)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Correlation Between Sepal & Petal Length')
plt.grid(True)


slope, intercept = np.polyfit(data['sepal_length'], data['petal_length'], 1)
abline_values = [slope * i + intercept for i in data['sepal_length']]
plt.plot(data['sepal_length'], abline_values, 
         linestyle = '-', 
         color = 'red')

plt.axvspan(5.5, 6.2, color = 'yellow', alpha = 0.2, linewidth = 0)
plt.show()
```
