```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

housing_df = pd.read_csv('real_estate_pricing.csv')
housing_df.head()
```

```python
matrix = np.triu(housing_df.corr())
plot = sns.heatmap(housing_df.corr(), annot = True, mask=matrix, cmap= 'coolwarm')
plt.show()
```

```python
plt.scatter(housing_df['Y house price of unit area'], housing_df['X4 number of convenience stores'], s=10)
plt.xlabel('House price per unit area')
plt.ylabel('Number of convenience stores')
plt.show()
```

```python
plt.scatter(housing_df['Y house price of unit area'], housing_df['X3 distance to the nearest MRT station'], s=10)
plt.xlabel('House price per unit area')
plt.ylabel('Distance to nearest MRT station')

price = housing_df['Y house price of unit area']
distance_station = housing_df['X3 distance to the nearest MRT station']

fit = np.polyfit(price, distance_station, 2)

plt.show()

```
